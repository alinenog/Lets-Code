```python
import requests as r
```

```python
#Declaração das fontes dos dados
url = 'https://api.covid19api.com/dayone/country/brazil'
resp = r.get(url)
```

```python
#Verificando resposta do servidor
resp.status_code
```

```python
#Guardando os dados retornando pela api,e vendo os dados
raw_data = resp.json()
raw_data[0]
```

```python
#Filtrar somente as informações que vamos ultilizar
final_data = []
for obs in raw_data:
  final_data.append([obs['Confirmed'], obs['Deaths'], obs['Recovered'], obs['Active'],obs['Date']])
```

```python
final_data
```

```python
final_data.insert(0, ['confirmados', 'obitos', 'recuperados', 'ativos', 'data'])
final_data
```

```python
CONFIRMADOS = 0
OBITOS = 1
RECUPERADOS = 2
ATIVOS = 3 
DATA = 4
```

```python
for i in range(1, len(final_data)):
    final_data[i][DATA] = final_data[i][DATA][:10]
```

```python
final_data
```

```python
import csv
import datetime as dt 
```

```python
with open('brasil-covid.csv' , 'w') as file:
    writer = csv.writer(file)
    writer.writerows(final_data)
```

```python
#Formatando a data /strptime passa o valor de string para data
for i in range(1, len(final_data)):
    final_data[i][DATA] = dt.datetime.strptime(final_data[i][DATA], '%Y-%m-%d')
```

```python
def get_datasets(y, labels):
    if type(y[0]) == list:
      datasets = []
      for i in range(len(y)):
          datasets.append({
            'label': labels[i],
            'data': y[i]
          })
      return datasets
    else:    
      return [
         {
           'label': labels[0],
           'data': y
         } 
      ]
```

```python
#titulo do grafico
def set_title(title=''):
    if title != '':
        display = 'true'  
    else:
      display = 'false'    
    return {
      'title': title,
      'display':display
    }
```

```python
#dicionario representativo dos nosso dados
def create_chart(x, y, labels, kind = 'bar', title= ''):

    datasets = get_datasets(y, labels)
    options = set_title(title)

    chart = {
      'type' : kind,
      'data' :{
        'labels' : x,
        'datasets' : datasets
      },
      'options': options
    }

    return chart
```

```python
#Função que vai fazer a requisião na api, usando esse dicionario

def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    resp = r.get(f'{url_base}?c={str(chart)}')
    return resp.content    
```

```python
#Função para salvar a imagem / 'wb' modo de escrita em binário
def save_image(path, content):
    with open(path, 'wb') as image:
      image.write(content)
```

```python
from PIL import Image
from IPython.display import display
```

```python
#Função para mostrar a imagem
def display_image(path):
    img_pil = Image.open(path)
    display(img_pil)
```

```python
#Plot de barras mostrando a evolução dos dados pulando de 10 em 10
y_data_1 = []
for obs in final_data[1::10]:
    y_data_1.append(obs[CONFIRMADOS])

y_data_2 = []
for obs in final_data[1::10]:
    y_data_2.append(obs[RECUPERADOS])    

labels = ['Confirmados', 'Recuperados']

x = []
for obs in final_data[1::10]:
    x.append(obs[DATA].strftime('%d/%m/%Y'))

#Criar grafico
chart = create_chart(x, [y_data_1, y_data_2], labels, title='Gráfico confirmados vs recuperados')
chart_content = get_api_chart(chart)
save_image('meu-primeiro-grafico.png', chart_content) 
display_image('meu-primeiro-grafico.png')   
```

```python
#Criando a função para gerar Qrc

def get_api_qrcode(link):
    text = quote(link) # parsing do link para url
    url_base = 'https://quickchart.io/qr'
    resp = r.get(f'{url_base}?text={text}')
    return resp.content
```
